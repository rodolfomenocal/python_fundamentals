```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/rodolfomenocal/Library/CloudStorage/OneDrive-InstitutoTecnologicoydeEstudiosSuperioresdeMonterrey/Documentos/Digital Hub/Python Fundamentals/athlete_events-1.csv")
df = df[df["Medal"].notnull()]
medallas_por_pais = df["Team"].value_counts()
medallas_por_pais = medallas_por_pais.head(10)
medallas_por_pais.plot(kind="bar", color="skyblue")
plt.show()
```

```python
ultimo_ao = df['Year'].max()
latest = df[df["Year"]== ultimo_ao]
medallas_ult_a = latest["Team"].value_counts()
medallas_ult_a.head(10).plot(kind="pie")
plt.show()
```
